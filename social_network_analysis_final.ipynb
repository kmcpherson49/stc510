```python
df = pd.DataFrame(data, index=['Nigeria','Northern Mali','Sahara','Algeria','Somalia','Sudan','Egypt','Libya','Tunisia','Palestine','Yemen','Saudi Arabia','Syria','Iraq','Chechnya','Pakistan','Afghanistan Pakistan','Afghanistan','Indonesia','Philippines'])
```

```python
df.index = df.index.astype(str)
df.columns = df.columns.astype(str)
```

```python
G = nx.Graph()
for source in df.index:
  for target in df.columns:
    if df.loc[source,target] == 1:
      G.add_edge(source, target)
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=50, node_color="blue", font_size=10, font_color="black", edge_color="gray", linewidths=1)
plt.title('Terrorist Allegences by Country')
plt.show()
```

From the graph you can see that there is no connection between groups across clusters meaning that groups are focused to their region instead of connected worldwide.

```python
nx.draw(G, pos, with_labels=True, node_size=50, node_color="blue", font_size=10, font_color="black", edge_color="red", linewidths=1)
plt.xlim(-1, 0.1)
plt.ylim(0.1, -1)
plt.title('Terrorist Allegences by Country- cluster close up')
plt.show()
```

You can see from these predominant cluster that they do have a central node to the group through which most edges pass. For example the cluster in the center left, Yemen acts as the connection for Dhamar governorate, Sana-a governorate, and Supporters of the Islamic State of Yemen.

If you were to remove any location nodes it would only affect the organizations in the area and not the data as a whole.

Overall I think any kind of research on terrorism moving forward so focus on specific regions as it seems that there is very little association outside of immediate location.
